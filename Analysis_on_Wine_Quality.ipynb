# **1. Introduction**
---



Wine producers constantly brag about the quality ratings that their wines receive from critics, because a high wine rating — implying high quality — translates into increased sales for a wine. But quality wines come in all colors, degrees of sweetness and dryness, and flavor profiles.

A wine’s quality is not absolute: how great a wine is or isn’t depends on who is doing the judging. The combined opinion of a group of trained, experienced palates (also known as wine experts) is usually considered a definitive judgment of a wine’s quality.

# Four Indicators of Wine Quality


-   **Complexity**

Higher quality wines are more complex in their flavor profile. They often have numerous layers that release flavors over time. Lower quality wines lack this complexity, having just one or two main notes that may or may not linger.

-   **Balance**

Wines that have good balance will be of higher quality than ones where one component stands out above the rest.

The five components – acidity, tannins, sugar/sweetness, alcohol and fruit – need to be balanced. For wines that need several years of aging to reach maturity, this gives them the time they need to reach optimal balance.

- **Typicity**

Another indicator of wine quality comes from typicity, or how much the wine looks and tastes the way it should.

For example, red Burgundy should have a certain appearance and taste, and it’s this combination that wine connoisseurs look for with each new vintage.

- **Intensity and Finish**

The final indicators of both white and red wine quality are the intensity and finish. High-quality wines will express intense flavors and a lingering finish, with flavors lasting after you’ve swallowed the wine. Flavors that disappear immediately can indicate that your wine is of moderate quality at best. The better the wine, the longer the flavor finish will last on your palate.

## **6.4 Handling outliers**


### **6.4.1 Identifying outliers**

We suspect that there are outliers in some of the variables. We will check each variable one by one.

```python
#let us define a function to identify outliers
#Code courtesy: Renu Khandelwal (https://medium.com/datadriveninvestor/finding-outliers-in-dataset-using-python-efc3fce6ce32)

def detect_outlier(data_1):
    
    outliers = []
    threshold = 3
    mean_1 = np.mean(data_1)
    std_1 =np.std(data_1)
    
    
    for y in data_1:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers
```

We will run the above function on each of the variables to identify outliers. We will then print the count of outliers, their minimum and maximum value. This will help us to identify whether the outliers are on the right side of the curve or left side.

```python
#fixed acidity
fixedacidity_list = winequality_df['fixed acidity'].tolist()

fixedacidity_outliers = detect_outlier(fixedacidity_list)
print(len(fixedacidity_outliers), min(fixedacidity_outliers), max(fixedacidity_outliers))
```

```python
#volatile acidity
volatileacidity_list = winequality_df['volatile acidity'].tolist()

volatileacidity_outliers = detect_outlier(volatileacidity_list)
print(len(volatileacidity_outliers), min(volatileacidity_outliers), max(volatileacidity_outliers))
```

    - This is the first major correlation with wine quality. Increase in alcohol percentage increases the wine quality
- **volatile acidity ==> {+} <== chlorides: 46%**
    - Both high amount of chlorides and volatile acidity is undesirable in wine. This may be the reason for positive correlation
- **free sulfur dioxide ==> {+} <== residual sugar: 43%**
    - This is similar to total sulfur dioxide

    - Acids are known to be denser than water. Increase in acids would increase density
